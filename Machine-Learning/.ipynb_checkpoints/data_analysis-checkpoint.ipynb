```python
# Initialize a list to store each DataFrame with gesture and id columns
all_data = []

# Iterate over files in the folder
for file_name in os.listdir(dataset_folder_path):
    if file_name.endswith('.csv'):
        # Extract gesture name from file name
        gesture_name = file_name.split('_')[0]
        
        # Read the CSV file into a DataFrame
        file_path = os.path.join(dataset_folder_path, file_name)
        df = pd.read_csv(file_path)
        
        # # Convert to Index
        # df.reset_index(drop=True, inplace=True)

        # Add gesture name and unique ID columns
        df['Gesture'] = gesture_name
        df['Id'] = file_name.split('_')[1].split('.')[0]  # ファイル名からIDを抽出（例: gesture1_1.csv の「1」部分）

        # Append to the list
        all_data.append(df)

# Concatenate all DataFrames into one
gesture_df = pd.concat(all_data, ignore_index=True)
```

### Mean acceleration

My assumption is <br>
- right   (+X)
- left    (-X)
- up      (-Z)
- down    (+Z)
- shuffle (X=0)
- circle  

```python
def plot_axis_means_from_df(df: pd.DataFrame) -> None:
    """
    Calculate the mean acceleration for X, Y, Z axes for each gesture's id,
    then calculate the overall mean for each gesture and plot the results in a bar chart.
    """
    # Step 1: Calculate mean for each gesture's id
    mean_by_id = df.groupby(['Gesture', 'Id'])[['X', 'Y', 'Z']].mean()
    
    # Step 2: Calculate overall mean for each gesture by averaging across ids
    mean_by_gesture = mean_by_id.groupby('Gesture').mean()

    # print(mean_by_id.head())
    # Extract data for plotting
    gesture_names = mean_by_gesture.index.tolist()
    x_means = mean_by_gesture['X'].tolist()
    y_means = mean_by_gesture['Y'].tolist()
    z_means = mean_by_gesture['Z'].tolist()

    
    # Plotting the results
    x = np.arange(len(gesture_names))  # the label locations
    width = 0.25  # the width of the bars

    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Plot each axis mean in a grouped bar chart
    ax.bar(x - width, x_means, width, label='X-axis')
    ax.bar(x, y_means, width, label='Y-axis')
    ax.bar(x + width, z_means, width, label='Z-axis')

    # Add labels, title, and legend
    ax.set_xlabel('Gesture')
    ax.set_ylabel('Mean Acceleration')
    ax.set_title('Mean Acceleration for X, Y, Z axes by Gesture')
    ax.set_xticks(x)
    ax.set_xticklabels(gesture_names)
    ax.legend()

    plt.tight_layout()
    plt.show()
```

### 3D Tracking

```python
def plot_random_3d_tracking_by_gesture(df: pd.DataFrame, col_name_list: list = ["X", "Y", "Z"]) -> None:
    """
    Randomly select and display data in 3D for each gesture's id from a DataFrame,
    highlighting the first and last points in red.
    """
    # Group by gesture
    for gesture_name, gesture_data in df.groupby('Gesture'):
        # Select one random id from each gesture group
        random_id = gesture_data['Id'].sample(n=1, random_state=42).iloc[0]
        selected_data = gesture_data[gesture_data['Id'] == random_id]
        
        # Set up the 3D plot for each gesture
        fig = plt.figure(figsize=(10, 8))
        ax = fig.add_subplot(111, projection='3d')
        
        # Plot the main trajectory in 3D
        ax.plot(selected_data[col_name_list[0]], 
                selected_data[col_name_list[1]], 
                selected_data[col_name_list[2]], 
                marker='o', color='b', alpha=0.7, linestyle='-', markersize=2)
        
        # Highlight the first and last points in red
        ax.scatter(selected_data[col_name_list[0]].iloc[0], 
                   selected_data[col_name_list[1]].iloc[0], 
                   selected_data[col_name_list[2]].iloc[0], 
                   color='red', s=50, label="Start")  # Starting point
        
        ax.scatter(selected_data[col_name_list[0]].iloc[-1], 
                   selected_data[col_name_list[1]].iloc[-1], 
                   selected_data[col_name_list[2]].iloc[-1], 
                   color='red', s=50, label="End")  # Ending point
        
        # Label the axes
        ax.set_xlabel(col_name_list[0])
        ax.set_ylabel(col_name_list[1])
        ax.set_zlabel(col_name_list[2])
        plt.title(f"3D Trajectory of Acceleration Data for Gesture: {gesture_name} (ID: {random_id})")
        ax.legend()
        
        plt.show()
```

### Waveform

```python
def display_waveform_overlay(df: pd.DataFrame, n_samples: int = 5,
                             col_name_list: list = ["X", "Y", "Z"]) -> None:
    """
    Display each gesture's data in an overlayed waveform plot.

    n_samples: Number of random points to sample and display for each gesture.
    """
    # 各ジェスチャーごとにプロットを生成
    gesture_names = df['Gesture'].unique()
    n_gestures = len(gesture_names)
    fig, axs = plt.subplots(n_gestures, 3, figsize=(15, 5 * n_gestures), sharex=True)
    
    # 各ジェスチャーごとに3列に分けて波形を重ねてプロット
    for row, gesture_name in enumerate(gesture_names):
        # 指定したサンプル数をランダムに選択
        gesture_data = df[df['Gesture'] == gesture_name]
        sampled_ids = gesture_data['Id'].drop_duplicates().sample(n=n_samples, random_state=42)
        
        for sample_id in sampled_ids:
            sample_data = gesture_data[gesture_data['Id'] == sample_id]
            index = sample_data.index
            
            # 各軸に重ねてプロット
            axs[row, 0].plot(sample_data[col_name_list[0]].values, alpha=0.7, label=f"{gesture_name}")
            axs[row, 1].plot(sample_data[col_name_list[1]].values, alpha=0.7, label=f"{gesture_name}")
            axs[row, 2].plot(sample_data[col_name_list[2]].values, alpha=0.7, label=f"{gesture_name}")

        # 各ジェスチャーの行ごとにラベルとタイトルを設定
        axs[row, 0].set_ylabel(f"{gesture_name.capitalize()} Acceleration")
        axs[row, 0].set_title(col_name_list[0])
        axs[row, 1].set_title(col_name_list[1])
        axs[row, 2].set_title(col_name_list[2])

        # Remove duplicate legends by keeping one per gesture
        handles, labels = axs[row, 0].get_legend_handles_labels()
        axs[row, 0].legend(handles[:1], labels[:1], loc="upper right")
    
    # プロットの設定
    for ax in axs.flat:
        ax.grid(True)
    
    plt.tight_layout()
    plt.show()
```

```python
display_waveform_overlay(gesture_df, n_samples=5)
```

### Fourier Transform

```python
def display_fft_from_df(df: pd.DataFrame) -> None:
    """
    Apply Fourier Transform to each gesture and display the frequency spectrum
    for X, Y, Z axes.
    """
    gesture_names = df['Gesture'].unique()
    
    for gesture_name in gesture_names:
        # 各ジェスチャーごとのデータを取得
        gesture_data = df[df['Gesture'] == gesture_name]
        
        # プロットの設定
        fig, axs = plt.subplots(1, 3, figsize=(15, 5))
        fig.suptitle(f"{gesture_name.capitalize()} Gesture - Fourier Transform")
    
        for axis, axis_name in enumerate(['X', 'Y', 'Z']):
            # フーリエ変換を適用するために、軸ごとのデータを取得
            combined_data = gesture_data[axis_name]
            
            # フーリエ変換を適用
            fft_result = np.fft.fft(combined_data)
            fft_magnitude = np.abs(fft_result)
            freq = np.fft.fftfreq(len(fft_result))
    
            # 周波数スペクトルをプロット（正の周波数のみ）
            axs[axis].plot(freq[:len(freq)//2], fft_magnitude[:len(fft_magnitude)//2])
            axs[axis].set_title(f"{axis_name}-axis Frequency Spectrum")
            axs[axis].set_xlabel("Frequency")
            axs[axis].set_ylabel("Amplitude")
    
        plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # タイトルと重ならないように余白を調整
        plt.show()
```

### One-hot Encoding

- Up      = 1
- Down    = 2
- Left    = 3
- Right   = 4
- Shuffle = 5
- Circle  = 6

```python
# add 0 for new columns
# gesture_df = gesture_df.assign(Up=0, Down=0, Left=0, Right=0, Shuffle=0, Circle=0)

gesture_encode = {"up": "Up", "down": "Down", "left": "Left", "right": "Right", "shuffle": "Shuffle", "circle": "Circle"}

def df_gesture_encode(df: pd.DataFrame, gesture_encode: dict)->pd.DataFrame:

    df = df.assign(Up=0, Down=0, Left=0, Right=0, Shuffle=0, Circle=0)
    
    for index, row in df.iterrows():
        gesture_name = row["Gesture"].lower()
        
        if gesture_name in gesture_encode:
            gesture_column = gesture_encode[gesture_name]
            df.at[index, gesture_column] = 1

    return df
```

### Outlier

```python
def check_outlier_by_mean(pd: pd.DataFrame) -> None:
    """
    Display mean x, y, z for each gesture and find outlier 
    """
    means_df = pd.groupby(['Gesture', 'Id']).agg(x_mean=('X', 'mean'), y_mean=('Y', 'mean'), z_mean=('Z', 'mean')).reset_index()
    
    # Get unique gestures to plot each on a separate row
    unique_gestures = means_df['Gesture'].unique()
    fig, axes = plt.subplots(len(unique_gestures), 3, figsize=(18, 5 * len(unique_gestures)))

    # If there's only one gesture, keep axes as a 2D list
    if len(unique_gestures) == 1:
        axes = [axes]

    for i, gesture in enumerate(unique_gestures):
        gesture_df = means_df[means_df['Gesture'] == gesture]
        
        # Scatter plot for x_mean vs y_mean with in-plot labels
        for _, row in gesture_df.iterrows():
            axes[i][0].scatter(row['x_mean'], row['y_mean'])
            axes[i][0].text(row['x_mean'], row['y_mean'], f"Id {row['Id']}", fontsize=9)
        
        axes[i][0].set_xlabel('x_mean')
        axes[i][0].set_ylabel('y_mean')
        axes[i][0].set_title(f'Scatter plot of x_mean vs y_mean for {gesture}')
        
        # Scatter plot for y_mean vs z_mean with in-plot labels
        for _, row in gesture_df.iterrows():
            axes[i][1].scatter(row['y_mean'], row['z_mean'])
            axes[i][1].text(row['y_mean'], row['z_mean'], f"Id {row['Id']}", fontsize=9)
        
        axes[i][1].set_xlabel('y_mean')
        axes[i][1].set_ylabel('z_mean')
        axes[i][1].set_title(f'Scatter plot of y_mean vs z_mean for {gesture}')
        
        # Scatter plot for x_mean vs z_mean with in-plot labels
        for _, row in gesture_df.iterrows():
            axes[i][2].scatter(row['x_mean'], row['z_mean'])
            axes[i][2].text(row['x_mean'], row['z_mean'], f"Id {row['Id']}", fontsize=9)
        
        axes[i][2].set_xlabel('x_mean')
        axes[i][2].set_ylabel('z_mean')
        axes[i][2].set_title(f'Scatter plot of x_mean vs z_mean for {gesture}')

    plt.tight_layout()
    plt.show()

check_outlier_by_mean(gesture_df)
```

### Normalize

```python
def normalize_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Normalize the X, Y, Z columns to the range [-1, 1] for each unique combination of Gesture and Id.
    """
    normalized_df = df.copy()

    normalized_df['X'] = normalized_df['X'].astype(float)
    normalized_df['Y'] = normalized_df['Y'].astype(float)
    normalized_df['Z'] = normalized_df['Z'].astype(float)
    
    for (gesture, gesture_id), group in df.groupby(['Gesture', 'Id']):
    
        x_min, x_max = group['X'].min(), group['X'].max()
        y_min, y_max = group['Y'].min(), group['Y'].max()
        z_min, z_max = group['Z'].min(), group['Z'].max()

        if x_max != x_min:
            normalized_df.loc[group.index, 'X'] = 2 * (group['X'] - x_min) / (x_max - x_min) - 1
        else:
            normalized_df.loc[group.index, 'X'] = 0  

        if y_max != y_min:
            normalized_df.loc[group.index, 'Y'] = 2 * (group['Y'] - y_min) / (y_max - y_min) - 1
        else:
            normalized_df.loc[group.index, 'Y'] = 0

        if z_max != z_min:
            normalized_df.loc[group.index, 'Z'] = 2 * (group['Z'] - z_min) / (z_max - z_min) - 1
        else:
            normalized_df.loc[group.index, 'Z'] = 0

    return normalized_df
```

### Remove Noise (Moving median)

```python
def get_moving_median_in_df(df: pd.DataFrame, data_num: int = 20) -> pd.DataFrame:
    """
    Divide each gesture's data into `data_num` chunks and calculate their median 
    values for X, Y, Z columns to reduce noise. Store these median values in new columns.
    """
    # Initialize an empty list to store intermediate DataFrames
    temp_dfs = []
        
    # Group by 'gesture' and 'id'
    for (gesture, gesture_id), group in df.groupby(['Gesture', 'Id']):

        # DEBUG:
        # print(f"Gesture: {gesture}, ID: {gesture_id}")
        
        # Segment size and remainder
        segment_size = len(group) // data_num
        remainder = len(group) % data_num

        # Lists to store median values for each segment and start times
        x_medians, y_medians, z_medians = [], [], []
        times = []

        # Calculate medians for each segment
        start_idx = 0
        for j in range(data_num):
            end_idx = start_idx + segment_size + (1 if j < remainder else 0)  # Distribute the remainder

            # Append median values for each segment
            x_medians.append(group['X'].iloc[start_idx:end_idx].median())
            y_medians.append(group['Y'].iloc[start_idx:end_idx].median())
            z_medians.append(group['Z'].iloc[start_idx:end_idx].median())
            times.append(group['Timestamp'].iloc[start_idx])  # Start time of the segment
            
            # Move to the next segment
            start_idx = end_idx

        # Create a DataFrame for the current gesture-id with median and time values
        temp_df = pd.DataFrame({
            'Timestamp': times,
            'Gesture': [gesture] * data_num,
            'Id': [gesture_id] * data_num,
            'X_median': x_medians,
            'Y_median': y_medians,
            'Z_median': z_medians
        })

        # Append the temporary DataFrame to the list
        temp_dfs.append(temp_df)

    # Concatenate all temporary DataFrames at once
    modified_df = pd.concat(temp_dfs, ignore_index=True)

    return modified_df
```

# Fourier Transform

```python
# Define the FFT_convert function to process the entire gesture_df with a loop internally
def FFT_convert_all(df: pd.DataFrame, num_of_freq: int = 50, 
                    col: list = ["X", "Y", "Z"]) -> pd.DataFrame:
    
    all_fft_data = []

    # Group by Gesture and Id, and apply FFT transformation for each group
    for (gesture, id_val), group in df.groupby(['Gesture', 'Id']):

        # Apply FFT to each axis (X, Y, Z) and select the first 50 frequencies
        fft_results = {
            'X_fft': np.abs(fft(group[col[0]]))[:50],
            'Y_fft': np.abs(fft(group[col[1]]))[:50],
            'Z_fft': np.abs(fft(group[col[2]]))[:50]
        }
            
        # Convert the FFT results into a DataFrame
        fft_df = pd.DataFrame(fft_results)
        
        # Add Gesture and Id columns
        fft_df['Gesture'] = gesture
        fft_df['Id'] = id_val

        # Copy other columns from the first row of the group, expanding to the length of fft_df
        for other_col in df.columns.difference(['X', 'Y', 'Z', 'Gesture', 'Id']):
            fft_df[other_col] = group[other_col].iloc[0]
    
        # Append the result to the list
        all_fft_data.append(fft_df)

    # Concatenate all FFT results into a single DataFrame
    final_fft_df = pd.concat(all_fft_data, ignore_index=True)
    
    return final_fft_df
```

```python
num_of_freq = 50

# convert no-noise data to FFT
col_name_list = ["X_median", "Y_median", "Z_median"]
gesture_df_median_fft = FFT_convert_all(gesture_df_median, 50, col_name_list)

# convert raw data to FFT
gesture_df_fft = FFT_convert_all(gesture_df)
gesture_df_fft
```

```python
# Fix the order of df
fft_df_col = ['Timestamp', 'Gesture', 'Id', 'X_fft', 'Y_fft', 'Z_fft', 'Up', 'Down', 'Left', 'Right', 'Shuffle', 'Circle']
raw_df_col = ['Timestamp', 'Gesture', 'Id', 'X', 'Y', 'Z', 'Up', 'Down', 'Left', 'Right', 'Shuffle', 'Circle']

def fix_col_order(df: pd.DataFrame, columns_order_list: list)->pd.DataFrame:
    return df.reindex(columns=columns_order_list)
```

```python
gesture_df_fft = fix_col_order(gesture_df_fft, fft_df_col)
gesture_df_median_fft = fix_col_order(gesture_df_median_fft, fft_df_col)
gesture_df = fix_col_order(gesture_df, raw_df_col)
```
